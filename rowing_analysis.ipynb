```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


lifting_example = pd.read_csv('lifting_dataset_1.csv')
heart_rate_example = pd.read_csv('heart_rate_dataset_1.csv')

# Plotting the lifting dataset correlations
plt.figure(figsize=(12, 6))
sns.regplot(data=lifting_example, x='compound_lift_weight_kg', y='average_watts')
plt.title('Correlation between Compound Lift Weight and Average Watts (Lifting Training)')
plt.xlabel('Compound Lift Weight (kg)')
plt.ylabel('Average Watts')
plt.grid(True)
plt.show()

# Plotting correlation between compound lift weight and 500m time
plt.figure(figsize=(12, 6))
sns.regplot(data=lifting_example, x='compound_lift_weight_kg', y='500m_time_sec')
plt.title('Correlation between Compound Lift Weight and 500m Sprint Time')
plt.xlabel('Compound Lift Weight (kg)')
plt.ylabel('500m Sprint Time (seconds)')
plt.grid(True)
plt.show()

# Plotting the heart rate dataset correlations
plt.figure(figsize=(12, 6))
sns.regplot(data=heart_rate_example, x='steady_state_hr_avg', y='average_watts')
plt.title('Correlation between Steady State HR and Average Watts (Heart Rate Training)')
plt.xlabel('Average Steady State Heart Rate')
plt.ylabel('Average Watts')
plt.grid(True)
plt.show()

# Plotting correlation between steady state HR and 500m time
plt.figure(figsize=(12, 6))
sns.regplot(data=heart_rate_example, x='steady_state_hr_avg', y='500m_time_sec')
plt.title('Correlation between Steady State HR and 500m Sprint Time')
plt.xlabel('Average Steady State Heart Rate')
plt.ylabel('500m Sprint Time (seconds)')
plt.grid(True)
plt.show()
```
